```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Coeffcients
alpha = 15.6
beta = 23
R = -1.143
C_2 = -0.714

#Create function that takes postion and time as input and returns differential equations as output
def chua(u, t):
    x, y, z = u
    # nonlinear resistor equation
    H = C_2*x + 0.5*(R-C_2)*(abs(x+1)-abs(x-1))
    #store 3 system of equations as chua
    system = [alpha*(y-x-H), x - y + z, -beta * y]
    return system

# time step
t_0 = 0
dt = 1e-3
t_final = 100
t = np.arange(t_0, t_final, dt)  #time step loop, starts at 0 and end at 100, with time step 1e-3

# initial conditions
u0 = [0.1,0,0]
# ode system instead
integrate = odeint(chua, u0, t)  #ode solver that takes chua equations, initial conditions, and time step

plot = plt.figure()
ax = plot.add_subplot(projection='3d') #3D plot
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Double scroll Attractor")

ax.plot(integrate[:,0],integrate[:,1],integrate[:,2])
```

```python
#LORENZ
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#function that takes xyz and coefficents as input and return the de's as output

def lorenz(xyz, *, s=10, r=14.9564, b=2.667):

 #Assign the differential equations and put them in an array

    x, y, z = xyz
    dx = s*(y - x)
    dy = r*x - y - x*z
    dz = x*y - b*z
    return np.array([dx, dy, dz])


dt = 0.01                                 #time step for loop
num_steps = 10000

xyzs = np.empty((num_steps + 1, 3))       # array with the steps and shape 3
xyzs[0] = (0., 1., 1.05)                  # Set initial values

# time step

for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt

# Plot
ax = plt.figure().add_subplot(projection='3d')

ax.plot(*xyzs.T, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

```python
#Rossler
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def rossler(xyz, *, a=0.2, b=0.2, c=12.6):

  #function that takes xyz and coefficents as input and return the de's as output
  #Assign the differential equations and put them in an array

    x, y, z = xyz
    dx = - y - z
    dy = x + a*y
    dz = b + z*(x-c)
    return np.array([dx, dy, dz])


dt = 0.01                                     #time step for loop
num_steps = 10000

xyzs = np.empty((num_steps + 1, 3))           # array with the steps and shape 3
xyzs[0] = (0., 1., 1)                         # Set initial values

#time step

for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + rossler(xyzs[i]) * dt

# 3D Plot
ax = plt.figure().add_subplot(projection='3d')

ax.plot(*xyzs.T, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Rossler Attractor")

plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Coeffcients
a = 0.2
b = 0.2
c = 15


#Create function that takes postion and time as input and returns differential equations as output
def rossler(u, t):
    x, y, z = u

    #store 3 system of equations as rossler
    system = [- y - z, x + a*y, b + z*(x-c)]
    return system

# time step
t_0 = 0
dt = 1e-3
t_final = 100
t = np.arange(t_0, t_final, dt)  #time step loop, starts at 0 and end at 100, with time step 1e-3

# initial conditions
u0 = [0.1,0,0]
# ode system instead
integrate = odeint(rossler, u0, t)  #ode solver that takes rossler equations, initial conditions, and time step

plot = plt.figure()
ax = plot.add_subplot(projection='3d') #3D plot
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Rossler attractor")

ax.plot(integrate[:,0],integrate[:,1],integrate[:,2])
```
